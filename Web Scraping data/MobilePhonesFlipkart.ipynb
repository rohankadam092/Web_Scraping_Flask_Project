```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
description = []
ram_storage = []
display = []
camera = []
battrey = []
price = []
rating = []
```

```python
pages = list(range(1,4))
for page in pages:
    url = requests.get("https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_ps&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=9114da71-3ddb-483f-b812-5ab3db58a9d2&p%5B%5D=facets.brand%255B%255D%3DIQOO&page={}".format(page)).text

    soup = BeautifulSoup(url,"html.parser")

    # description
    desc = soup.find_all("div", class_="_4rR01T")
    for i in range (0,len(desc)):
        description.append(desc[i].text)


    commonclass = soup.find_all("li", class_="rgWa7D")
    for i in range (0,len(commonclass)):
        com = commonclass[i].text

        # ram/storage
        if ("RAM" in com or "ROM" in com):
            ram_storage.append(com)

        # display
        elif ("Display" in com):
            display.append(com)

        # camera
        elif ("Camera" in com):
            camera.append(com)

#         # battrey
        elif ("Battery" in com):
            battrey.append(com)

    # price
    pri = soup.find_all("div", class_="_30jeq3 _1_WHN1")
    for i in range (0,len(pri)):
        price.append(pri[i].text)

    # rating
    rat = soup.find_all("div", class_="_3LWZlK")
    for i in range (0,len(rat)):
        rating.append(rat[i].text)
```

```python
print(len(description))
print(len(ram_storage))
print(len(display))
print(len(camera))
print(len(battrey))
print(len(price))
print(len(rating))
```

```python
data = {"_4rR01T":description[slice(51)],"RAM":ram_storage[slice(51)],"Display":display[slice(51)],
        "Camera":camera[slice(51)],"Battery":battrey[slice(51)],
        "_30jeq3 _1_WHN1":price[slice(51)]}
#         "_3LWZlK":rating[slice(146)]}
mobilephone = pd.DataFrame(data)
mobilephone
```
