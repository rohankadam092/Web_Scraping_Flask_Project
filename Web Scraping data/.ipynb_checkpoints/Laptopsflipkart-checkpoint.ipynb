```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
description = []
processor = []
ram = []
os = []
storage = []
display = []
price = []
```

```python
pages = list(range(1,20))
for page in pages:
    url = requests.get("https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DDELL&p%5B%5D=facets.brand%255B%255D%3Dacer&p%5B%5D=facets.brand%255B%255D%3DMSI&p%5B%5D=facets.brand%255B%255D%3DAPPLE&p%5B%5D=facets.brand%255B%255D%3DMi&p%5B%5D=facets.brand%255B%255D%3DAvita&p%5B%5D=facets.brand%255B%255D%3DREDMI&p%5B%5D=facets.brand%255B%255D%3DInfinix&p%5B%5D=facets.brand%255B%255D%3DLG&p%5B%5D=facets.brand%255B%255D%3DVaio&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DALIENWARE&p%5B%5D=facets.brand%255B%255D%3Drealme&p%5B%5D=facets.brand%255B%255D%3DMICROSOFT&p%5B%5D=facets.brand%255B%255D%3DSmartron&p%5B%5D=facets.brand%255B%255D%3Diball&p%5B%5D=facets.brand%255B%255D%3DNokia&p%5B%5D=facets.brand%255B%255D%3DMicromax&page={}".format(page)).text

    soup = BeautifulSoup(url,"html.parser")
    
    # description
    desc = soup.find_all("div", class_="_4rR01T")
    for i in range (0,len(desc)):
        description.append(desc[i].text)


    commonclass = soup.find_all("li", class_="rgWa7D")
    for i in range (0,len(commonclass)):
        com = commonclass[i].text

        # processor
        if ("Processor" in com):
            processor.append(com)

        # ram
        elif ("RAM" in com):
            ram.append(com)

        # os
        elif ("Operating" in com):
            os.append(com)

        # storage
        elif ("SSD" in com or "HDD" in com):
            storage.append(com)
        
        #display
        elif ("Display" in com):
            display.append(com)
        

    # price
    pri = soup.find_all("div", class_="_30jeq3 _1_WHN1")
    for i in range (0,len(pri)):
        price.append(pri[i].text)
```

```python
print(len(description))
print(len(processor))
print(len(ram))
print(len(os))
print(len(storage))
print(len(display))
print(len(price))
```

```python
data = {"Description":description[slice(447)],"Processor":processor[slice(447)],"RAM":ram[slice(447)],
        "Operating System":os[slice(447)],"Storage":storage[slice(447)],"Display":display[slice(447)],
        "Price":price[slice(447)]}
laptop = pd.DataFrame(data)
laptop
```

```python
laptop.to_csv("laptop_flipkart_all.csv")
```
